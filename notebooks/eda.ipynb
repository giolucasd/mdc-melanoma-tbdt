```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="white", context="talk")

palette = {
    "Treino": "#4C72B0",
    "Validação": "#55A868",
    "Teste": "#C44E52"
}
```

```python
test_df = pd.read_csv("../data/sample_submission.csv")
train_df = pd.read_csv("../data/train.csv")
val_df = pd.read_csv("../data/val.csv")
```

```python
counts = pd.DataFrame({
    "Split": ["Treino", "Validação", "Teste"],
    "Quantidade": [len(train_df), len(val_df), len(test_df)]
})

plt.figure(figsize=(7, 5))
ax = sns.barplot(
    data=counts,
    x="Split",
    y="Quantidade",
    palette="pastel"
)

# remover bordas desnecessárias
sns.despine()

# valores em cima das barras
for p in ax.patches:
    ax.annotate(
        f"{int(p.get_height())}",
        (p.get_x() + p.get_width() / 2, p.get_height()),
        ha="center",
        va="bottom",
        fontsize=12
    )

# ax.set_title("Distribuição dos dados por conjunto")
ax.set_xlabel("")
ax.set_ylabel("Imagens")

plt.tight_layout()
plt.show()
```

```python
def plot_pie(df, title):
    counts = df["target"].value_counts().sort_index()
    labels = ["Negativo", "Positivo"]
    colors = ["#DD8452", "#4C72B0"]

    plt.figure(figsize=(5, 5))
    plt.pie(
        counts,
        labels=labels,
        autopct="%1.1f%%",
        startangle=90,
        colors=colors,
        wedgeprops={"edgecolor": "white"}
    )
    plt.title(title)
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd

data = {
    "Configuração": [
        "Cabeça",
        "Cabeça + 1 camada",
        "Cabeça + 2 camadas",
        "Cabeça + 4 camadas",
        "Cabeça + tudo",
    ],
    "Score": [0.746, 0.782, 0.817, 0.824, 0.857],
}

df = pd.DataFrame(data)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid", context="talk")

plt.figure(figsize=(8, 4))

ax = sns.barplot(
    data=df,
    x="Score",
    y="Configuração",
    palette="viridis"
)

# Anotar valores nas barras
for i, v in enumerate(df["Score"]):
    ax.text(v + 0.002, i, f"{v:.3f}", va="center")

ax.set_xlim(0.7, 0.9)
ax.set_xlabel("Score")
ax.set_ylabel("")
ax.set_title("Impacto do Descongelamento do Encoder")

plt.tight_layout()
plt.show()
```
